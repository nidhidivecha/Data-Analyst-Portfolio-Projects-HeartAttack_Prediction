### Correlation matrix using corr() method as a Heatmap 

This is basically used to find out if all the features are positively or negatively correlated.

The correlation coefficient is a value in the matrix between -1 and 1

A value close to 1 indicates a strong positive relationship, 
a value close to -1 indicates a strong negative relationship, 
and a value close to 0 indicates no relationship.

```python
corr_matrix = df.corr().round(2)
```

```python
# Plotting correlation_matrix as a heatmap using seaborn 
# (annot = True) parameter displays correlation coefficient 
# cmap parameter - sets color of the heatmap

plt.subplots(figsize=(13,10))
# sns.heatmap(corr_matrix, annot=True, cmap=sns.diverging_palette(220, 20, as_cmap=True))
# sns.heatmap(corr_matrix, annot=True, cmap=sns.color_palette("crest"))
# sns.heatmap(corr_matrix, annot=True, cmap=sns.color_palette("flare"))
# sns.heatmap(corr_matrix, annot=True, cmap=sns.color_palette("rocket"))
sns.heatmap(corr_matrix, annot=True, cmap=sns.cubehelix_palette(as_cmap=True))
```

**We can observe that, there is positive correlation between cp(chest pain) and target variable i.e as cp increases there are more chances of having heart attacks**

**On the contrary, we see a negative correlation between  exercise induced angina (exang) and target variable. This makes sense because when you excercise, you have less risk of having a HeartDisease**

### DATA PREPROCESSING

```python
# Choosing features and target labels. y: target (independent feature), x: features (dependent feature)

x=df.drop(['target'],axis=1)
y=df['target']
```

#### Separate the dataframes into x_train, x_test, y_train, and y_test

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3)
```

```python
x_train.shape[0], x_test.shape[0]
```

```python
#Normalize: Standardizing the data will transform the data so that its distribution will have a 
#mean of 0 and a standard deviation of 1

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

# Data Modelling 

Build a baseline model to predict using a Logistic Regression and explore the results.


```python

from sklearn.linear_model import LogisticRegression 
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix, accuracy_score
```

###  Model -1 Logistic Regession

```python
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)

print(f"The accuracy score is: {accuracy_score(y_test,y_pred)}")
print("-------------------------------------------------------------------------")
print(f"The Confusion Matrix is:  \n{confusion_matrix(y_test,y_pred)}")
print("-------------------------------------------------------------------------")
print(f"The Classification Report is: \n{classification_report(y_test,y_pred)}")
```

### Model - 2 Decision tree classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)

print(f"The accuracy score is: {accuracy_score(y_test,y_pred)}")
print("-------------------------------------------------------------------------")
print(f"The Confusion Matrix is:  \n{confusion_matrix(y_test,y_pred)}")
print("-------------------------------------------------------------------------")
print(f"The Classification Report is: \n{classification_report(y_test,y_pred)}")
```

### Model - 3 Random Forest

```python
model = RandomForestClassifier(n_estimators = 10)
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)

print(f"The accuracy score is: {accuracy_score(y_test,y_pred)}")
print("-------------------------------------------------------------------------")
print(f"The Confusion Matrix is:  \n{confusion_matrix(y_test,y_pred)}")
print("-------------------------------------------------------------------------")
print(f"The Classification Report is: \n{classification_report(y_test,y_pred)}")
```

### Model - 4 KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors = 12)
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)

print(f"The accuracy score is: {accuracy_score(y_test,y_pred)}")
print("-------------------------------------------------------------------------")
print(f"The Confusion Matrix is:  \n{confusion_matrix(y_test,y_pred)}")
print("-------------------------------------------------------------------------")
print(f"The Classification Report is: \n{classification_report(y_test,y_pred)}")
```

### Model - 5 SVM

```python
from sklearn.svm import SVC
```

```python
model = SVC()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)

print(f"The accuracy score is: {accuracy_score(y_test,y_pred)}")
print("-------------------------------------------------------------------------")
print(f"The Confusion Matrix is:  \n{confusion_matrix(y_test,y_pred)}")
print("-------------------------------------------------------------------------")
print(f"The Classification Report is: \n{classification_report(y_test,y_pred)}")
```

```python
algorithms = [
    LogisticRegression(),
    DecisionTreeClassifier(),
    RandomForestClassifier(n_estimators=10),
    KNeighborsClassifier(n_neighbors=12),
    SVC()
]
model_names = [
    "LogisticRegression",
     "DecisionTreeClassifier",
     "RandomForestClassifier",
     "KNeighborsClassifier",
     "SVC"
]
```

```python
results = []
for model, model_name in zip(algorithms, model_names):
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    accuracy = accuracy_score(y_test, y_pred)
    results.append({"Algorithm": model_name, "Accuracy": accuracy})
report = pd.DataFrame(results)
report.sort_values("Accuracy", ascending=False)
```

*Here the best performing model is KNN with a accuraccy score of 85%*
